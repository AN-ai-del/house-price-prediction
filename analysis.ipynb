# Task 2: Data Cleaning and Preprocessing

The dataset was inspected for missing values and duplicate records.

Since no missing values or duplicate rows were found, no rows were removed during the cleaning process.

The next step is to prepare categorical features for machine learning models.

```python
# Create a copy of the dataset

housing_df = df.copy()

print("Working dataset created successfully.")
```

```python
# Verify missing values

missing_values = housing_df.isnull().sum()

print("Total missing values:")
print(missing_values.sum())
```

```python
# Verify duplicate rows

duplicate_count = housing_df.duplicated().sum()

print(f"Duplicate rows found: {duplicate_count}")
```

```python
# Binary categorical columns

binary_columns = [
    "mainroad",
    "guestroom",
    "basement",
    "hotwaterheating",
    "airconditioning",
    "prefarea"
]

# Convert yes/no to 1/0

for column in binary_columns:
    housing_df[column] = housing_df[column].map({
        "yes": 1,
        "no": 0
    })
```

```python
# Verify binary encoding

housing_df[binary_columns].head()
```

```python
# One-hot encode furnishing status

housing_df = pd.get_dummies(
    housing_df,
    columns=["furnishingstatus"],
    drop_first=True,
    dtype=int
)
```

```python
# Display updated columns

housing_df.columns.tolist()
```

```python
# Dataset information after preprocessing

housing_df.info()
```

```python
# Define target variable

target_column = "price"

# Create feature matrix and target vector

X = housing_df.drop(columns=[target_column])

y = housing_df[target_column]

print(f"Feature matrix shape: {X.shape}")
print(f"Target vector shape: {y.shape}")
```

## Preprocessing Summary

The dataset required minimal cleaning because no missing values or duplicate records were present.

Binary categorical variables were converted from yes/no format into numerical values (1/0).

The furnishing status column was transformed using one-hot encoding to make it suitable for machine learning models.

The dataset is now fully prepared for model training and evaluation.
